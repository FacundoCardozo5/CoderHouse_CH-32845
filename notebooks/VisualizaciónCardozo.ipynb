### Cada una de las regiones tiene 338 entradas. El siguiente paso es elegir una región, para la cual nos gustaría poder predecir el precio futuro de las paltas

```python
PREDICTING_FOR = "TotalUS"
```

```python
date_price = regions.get_group(PREDICTING_FOR)[['Date', 'AveragePrice']].reset_index(drop=True)
```

```python
date_price.plot(x='Date', y='AveragePrice', kind="line")
```

# Scatter Plot

```python
plt.scatter(df['year'], df['AveragePrice'])
plt.xlabel('Año')
plt.ylabel('Precio Promedio')
```

```python
data1 = df['Total Volume'].value_counts(). sort_values(ascending=False).head(5)
data1 = pd.DataFrame(data1)
data1 = data1.reset_index()
sns.set()
plt.figure(figsize=(20,6))
sns.barplot(x='index', y='Total Volume', data=data1, palette='RdBu')
```

```python
plt.figure(figsize=(12,20))
sns.set_style('whitegrid')
sns.pointplot(x='AveragePrice', y='region', data=df, hue='type', join=False)
plt.xlabel('region', {'fontsize' : 'large'})
plt.ylabel('AveragePrice', {'fontsize' : 'large'})
plt.title("Precio promedio por region", {'fontsize' : 10})
```
